```python
# pandas 1.4.2
import pandas as pd

# numpy 1.22.3
import numpy as np

# matplotlib 3.5.1
import matplotlib.pyplot as plt
import matplotlib.animation as animation
# %matplotlib inline
%matplotlib ipympl

import math

plt.rcParams['figure.figsize'] = [10, 5]

%reload_ext autoreload
%autoreload 2
```

```python
def x(t):
    return t


def y(t):
    return math.sin(t)


def dx(t):
    return 1


def dy(t):
    return math.cos(t)


def norm(vec):
    return math.sqrt(sum([coord**2 for coord in vec]))
```

```python
ts = np.linspace(0, 5*np.pi, 300)

df_coord = pd.DataFrame({
    't': ts,
    'x': [x(t) for t in ts],
    'y': [y(t) for t in ts],
    'dx': [dx(t) for t in ts],
    'dy': [dy(t) for t in ts],
})

df_coord.head()
```

```python
fig, ax = plt.subplots()

l = plt.plot(df_coord.x,df_coord.y)

ax = plt.axis([0, 5*np.pi, -1, 1])

redDot, = plt.plot([0], [np.sin(0)], 'ro')


def animate(i):
    redDot.set_data(i, np.sin(i))
    return redDot,


myAnimation = animation.FuncAnimation(
    fig,
    animate,
    frames=np.linspace(0, 5*np.pi, 300),
    interval=10,
    blit=True,
    repeat=True,
)

plt.show()
```
