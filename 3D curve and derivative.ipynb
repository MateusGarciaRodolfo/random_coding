```julia
using Plots
```

```julia
gr()
```

```julia
γ(t) = (cos(5t), sin(5t), t)

norm(vec) = sqrt(sum([coord^2 for coord in vec]))

dγ_dt(t) = [-5sin(5t), 5cos(5t), 1]
dγ_norm(t) = dγ_dt(t)/norm(dγ_dt(t))
```

```julia
ts = range(0, pi, length=100)
plot([γ(t) for t in ts], label=false)
```

```julia
plot([γ(t) for t in ts], label=false)
t = 1
vec_start = γ(t)
vec_end = dγ_norm(t)
plot!([vec_start[1], vec_start[1] + vec_end[1]], [vec_start[2], vec_start[2] + vec_end[2]], [vec_start[3], vec_start[3] + vec_end[3]], arrow=true)
```

```julia
plt_list = []

ts = range(0, pi, length=100)

for t in ts
    vec_start = γ(t)
    vec_end = dγ_norm(t)

    p = plot([γ(t) for t in ts], label=false, xlims=(-1.5, 1.5), ylims=(-1.5, 1.5), zlims=(0, 3.5), xformatter=_ -> "", yformatter=_ -> "", zformatter=_ -> "")
    plot!([vec_start[1], vec_start[1] + vec_end[1]], [vec_start[2], vec_start[2] + vec_end[2]], [vec_start[3], vec_start[3] + vec_end[3]], arrow=true, label=false)

    push!(plt_list, p)
end
```

```julia
anim = Animation()

for plt in plt_list
    frame(anim, plt)
end
```

```julia
gif(anim,
    # "derivative_parametric.gif",
    fps=20,
    show_msg=false)
```
